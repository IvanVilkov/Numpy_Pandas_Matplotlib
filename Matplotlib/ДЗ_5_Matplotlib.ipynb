### **Задача 2**

Используя библиотеку Matplotlib, постройте графики функций $y = x^2$, $y = -x^2$ и $y = sin(1/x)*x^2$ на одном графике на промежутке от -1 до 1. 

* Дайте название графику 
* Дайте названия осям
* Присвойте лейблы и выведите легенду
* Графики функций должны быть отрисованы пунктирными линиями (любыми из доступных на ваш вкус) 

```python
# Построение 3-х графиков в одном
x=np.linspace(-1,1,100)
plt.figure(figsize=(9,7))
plt.plot(x,x**2,'r-.', label='y=x**2')
plt.plot(x,-x**2, 'b--', label='y=-x**2')
plt.plot(x,np.sin(1/x)*x**2, 'k:',label='y=sin(1/x)*x**2', linewidth=3)
plt.title('Functions')
plt.xlabel('ось х')
plt.ylabel('ось у')
plt.legend()
plt.show()
```

На графиках выводите "распределение признаков", т.е. зависимость признака от его индекса

```python
from sklearn import datasets
iris = datasets.load_iris(return_X_y=False)
data = iris.data
target = iris.target
names = iris.target_names
feature=iris.feature_names
df=pd.DataFrame(data=data, columns=feature)
df['target']=target
df['names']=names[0]
df.loc[target==1,'names']=names[1]
df.loc[target==2,'names']=names[2]
df
```

```python

plt.figure(figsize=(15,10))
plt.subplot(221)
plt.scatter(data=df, x=df.index, y=df.columns[0], c=df.columns[4])
plt.title(df.columns[0])

plt.subplot(222)
plt.scatter(data=df, x=df.index, y=df.columns[1], c=df.columns[4])
plt.title(df.columns[1])

plt.subplot(223)
plt.scatter(data=df, x=df.index, y=df.columns[2], c=df.columns[4])
plt.title(df.columns[2])

plt.subplot(224)
plt.scatter(data=df, x=df.index, y=df.columns[3], c=df.columns[4])
plt.title(df.columns[3])

plt.show()
```
