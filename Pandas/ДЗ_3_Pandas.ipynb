Самостоятельно изучите метод df.pivot и с помощью него в 1 строку получите новую таблицу, где вдоль строки будут именами друзей, колонки будут названиям фруктов, а на пересечении строк и столбцов будет цена, за которую был куплен фрукт кем-то из друзей.
````python
# имена ['Петя', 'Вася', 'Маша', 'Даша', 'Маша', 'Вася', 'Маша']
# фрукты ['яблоко', 'груша', 'персик', 'абрикос', 'яблоко', 'персик', 'груша']
# цены [50, 40, 23, 37, 66, 3, 15]
````
